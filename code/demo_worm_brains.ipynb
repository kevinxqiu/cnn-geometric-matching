# WormBrain : Demo on how to apply a transformation to worm brain images

Given a source worm brain image, it's aligned to a target worm brain image applying an affine and a tps transformation.

```python
from functions.model_alignment import * #load models, save_warped, etc.
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 
warnings.filterwarnings("ignore", category=FutureWarning) 
```

## Load models

```python
#load a model with the best weights we have trained on worm brain images. 
#the weights are loaded in the best architecture we have
w_aff = './trained_models/wormbrain_simplerfr/affine_wormbrain_simplerfr.pth.tar'
w_tps = './trained_models/wormbrain_simplerfr/tps_wormbrain_simplerfr.pth.tar'

model_aff, model_tps = load_model(aff_params_path=w_aff, aff_feat_ext = 'wormbrain_1', aff_feat_reg = 'simpler',
                                  tps_params_path=w_tps, tps_feat_ext = 'wormbrain_1', tps_feat_reg = 'simpler')
```

## Warp images

```python
#we choose two example images, the source is aligned with the target one. This task is done in two steps, affine and tps, as shown below.
#the final result is generated with the combination of both transformations
source_image = 'datasets/rgb512_augmented/TrainVal/frame_87.jpg'
target_image = 'datasets/rgb512_augmented/TrainVal/frame_97.jpg'

save_warped(source_name=source_image, target_name=target_image, savename='', model_aff=model_aff, model_tps=model_tps, demo=True)
```
